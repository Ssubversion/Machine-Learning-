```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(['adult_male', 'deck', 'embark_town',
       'alive','who','class'],axis=1,inplace=True)
```

```python
df.info()
```

fill the missing values in age and embarked

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

889 values in embarked 2 values are null either fill null values using mean , median , mode or remove 2 rows of embarked column so all columns will remove the 2 rows total will be 889

```python
df.dropna(subset=['embarked'],inplace=True)
```

```python
df.info()
```

```python
df.head()
```

we have to predict person survived or not

# LABEL ENCODING

we will label encode - sex,embarked

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
X=df.drop('survived',axis=1)
y=df['survived']
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
