# ENCODING - hot and label to check whos best

```python
df.columns
```

```python
X_one_encode = pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=True)
'''all columns not having int values'''
```

```python
X_one_encode
```

```python
X_one_encode=X_one_encode.astype(int)
```

```python
X_one_encode
```

lets do label encoding too

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
columns = ['model','transmission','fuelType']
X_label = X
for i in columns:
    X_label[i]=encode.fit_transform(X_label[i])
```

```python
X_label
```

label encoding dont create extra columns
now lets check model performance using one hot encoding and label encoding 

# STANDARD SCALING 


```python
from sklearn.preprocessing import StandardScaler
```

scaling all numerical columns - year,mileage,tax,mpg,enginesize - other things are encoded already

```python
numerical_cols = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()
X_one_encode[numerical_cols] = scaler.fit_transform(X_one_encode[numerical_cols])
```

```python
X_one_encode
```

now scaling for label encoding too

```python
X_label.columns
```

```python
X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
X_label
```

# MODEL CREATION USING ONE HOT ENCODING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X_one_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

adjusted r2

```python
n= X_test.shape[0]
p= X_test.shape[1]
adjusted_r2 = 1-((1-r2)*(n-1))/(n-p-1)
adjusted_r2
```

# MODEL CREATION USING LABEL ENCODING

```python
 X_train, X_test, y_train, y_test = train_test_split(X_label, y, test_size=0.33, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

one hot encoding model performed better
